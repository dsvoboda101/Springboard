```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

## Understand the Open-Meteo API

- **API Endpoint**: `https://archive-api.open-meteo.com/v1/archive`
- **Parameters**:
  - `latitude`: Latitude of the location.
  - `longitude`: Longitude of the location.
  - `start_date`: Start date of the data (YYYY-MM-DD).
  - `end_date`: End date of the data (YYYY-MM-DD).
  - `hourly`: Comma-separated list of variables (e.g., temperature, precipitation).
  - `timezone`: Timezone of the data.

**Documentation**: [Open-Meteo API Docs](https://open-meteo.com/en/docs)

```python
# Step 1: Set Up Parameters

# Location coordinates for New York City
latitude = 40.7128
longitude = -74.0060

# Date range
start_date = '2023-01-01'
end_date = '2023-01-07'  # One week of data

# Variables to fetch
hourly_variables = ['temperature_2m', 'relativehumidity_2m', 'precipitation']
```

```python
# Step 2: Make the API Request

base_url = 'https://archive-api.open-meteo.com/v1/archive'

params = {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/New_York'
}

response = requests.get(base_url, params=params)
```

```python
# Check if the request was successful

if response.status_code == 200:
    print('Data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response.status_code}')
```

```python
# Step 3: Load Data into pandas DataFrame

data = response.json()
hourly_data = data['hourly']

df = pd.DataFrame(hourly_data)
df.head()
```

```python
# Step 4: Data Cleaning

# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set 'time' as the index
df.set_index('time', inplace=True)

# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
df.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
# Step 5: Exploratory Data Analysis

# Summary Statistics
df.describe()
```

```python
# Plot Temperature Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['temperature_2m'], label='Temperature (C)')
plt.title('Temperature Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()
```

```python
# Plot Relative Humidity Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['relativehumidity_2m'], color='orange', label='Relative Humidity (%)')
plt.title('Relative Humidity Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

```python
# Plot Precipitation Over Time

plt.figure(figsize=(14, 6))
plt.bar(df.index, df['precipitation'], label='Precipitation (mm)')
plt.title('Precipitation Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Step 6: Correlation Analysis

# Calculate correlation between temperature and humidity
correlation = df['temperature_2m'].corr(df['relativehumidity_2m'])
print(f'Correlation between Temperature and Relative Humidity: {correlation:.2f}')
```

```python
# Scatter Plot of Temperature vs. Relative Humidity

plt.figure(figsize=(8, 6))
plt.scatter(df['temperature_2m'], df['relativehumidity_2m'], alpha=0.5)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# Step 7: Resample Data to Daily Averages

daily_avg = df.resample('D').mean()
daily_avg.head()
```

```python
# Plot Daily Average Temperature

plt.figure(figsize=(10, 5))
plt.plot(daily_avg.index, daily_avg['temperature_2m'], marker='o')
plt.title('Daily Average Temperature in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.grid(True)
plt.show()
```

```python
# Step 8: Compare with Another Location (Los Angeles)

# Coordinates for Los Angeles
latitude_la = 34.0522
longitude_la = -118.2437

params_la = {
    'latitude': latitude_la,
    'longitude': longitude_la,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_la = requests.get(base_url, params=params_la)

# Load and clean Los Angeles data
data_la = response_la.json()
hourly_data_la = data_la['hourly']
df_la = pd.DataFrame(hourly_data_la)

df_la['time'] = pd.to_datetime(df_la['time'])
df_la.set_index('time', inplace=True)
df_la.fillna(method='ffill', inplace=True)
```

```python
# Resample Los Angeles Data to Daily Averages

daily_avg_la = df_la.resample('D').mean()
```

```python
# Combine DataFrames for Comparison

combined_temp = pd.DataFrame({
    'New York': daily_avg['temperature_2m'],
    'Los Angeles': daily_avg_la['temperature_2m']
})
```

```python
# Plot Comparison of Daily Average Temperature

combined_temp.plot(kind='bar', figsize=(10, 6))
plt.title('Daily Average Temperature: New York vs Los Angeles')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Step 9: Advanced Analysis - Temperature Difference

# Calculate Temperature Difference
combined_temp['Temp Difference'] = combined_temp['Los Angeles'] - combined_temp['New York']
combined_temp
```

```python
# Step 10: Save Data to CSV (Optional)

# Save New York data
df.to_csv('new_york_weather.csv')

# Save Los Angeles data
df_la.to_csv('los_angeles_weather.csv')
```

```python
# Plot Precipitation Over Time - With resampling

plt.figure(figsize=(14, 6))
plt.plot(combined_weekly_avg.index, combined_weekly_avg['London'], label='London')
plt.plot(combined_weekly_avg.index, combined_weekly_avg['Seattle'], label='Seattle')
plt.title('Precipitation Over Time in London and Seattle(Weekly Averages)')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

Weekly averages are easier to interpret, and confirm that Seattle has consistently higher rain over most othe year.  Both cities are relatively dry in the summer.  The most stricking difference is that Seattle is VERY wet in the winter.

```python
#Group originial houly data by the total prepcipitation per month in a new data frame called 'monthly_sums'. 
combined_precip['Month'] = combined_precip.index.month_name()
combined_precip['Month_Num'] = combined_precip.index.month
monthly_sums = combined_precip.groupby('Month_Num')[['London', 'Seattle']].sum().reset_index() #Using Month_Num keeps the months in chonological order
monthly_sums['Month'] = monthly_sums['Month_Num'].apply(lambda x: pd.Timestamp(2024, x, 1).month_name())
monthly_sums = monthly_sums.set_index('Month')
monthly_sums = monthly_sums[['London', 'Seattle']]
monthly_sums.head()
```

Plot the data by the total precipitation per month to get a better idea of the net differences in precipitiaton by month between the two cities.

```python
# Plot Total rain per month in 2022 
monthly_sums.plot(kind='bar', figsize=(10, 6))
plt.title('Precipitation in 2022 in London and Seattle (Total per month)')
plt.xlabel('Month')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

To further illustrate the relative size of the differences in precipitation between Seattle and London by month, make a new data frame called "monthly_differece" that shows the difference in precipitation as Seattle - London.

```python
monthly_sums['difference (S-L)'] = monthly_sums['Seattle'] - monthly_sums['London']
monthly_difference = monthly_sums[['difference (S-L)']]
monthly_difference.head(12)
```

Plot the difference in precipitation by month, using difference colors to show for months when Seattle or London have more precipitation.

```python
# Plot Total rain per month in 2022 - With resampling
colors = ['blue' if val >= 0 else 'red' for val in monthly_difference.iloc[:, 0]]
values = monthly_difference.iloc[:, 0]
# Plot
ax = values.plot(
    kind='bar',
    figsize=(10, 6),
    color=colors,
    legend=False  # remove default legend
)

# Add solid black line at y = 0
ax.axhline(0, color='black', linewidth=1.5)

# Labels and title
plt.title('Difference in Monthly Precipitation between Seattle and London in 2022')
plt.xlabel('Month')
plt.ylabel('Precipitation (mm): Seattle - London')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Conclusion: Seattle had more rain that London in 2022.  This is primarily due to Seattle having a very rainy winter, while London doesn't seem to have a distinct "rainy" season. The greatest difference in rainfall was in January through April, and again in December, with the largest differences in January (Seattle had 190.9 mm more precipitation than London) and December (Seattle had 135.2 mm more precipitation than London).  The trend reversed in July through November (although to a much smaller degree) when Seattle is having a dry season (July through September), when London had similar or more precipitation than Seattle during these months.  The month were London had the biggest difference in precipiation over Seattle was September, with London having 40 more mm of precipitation than Seattle.
